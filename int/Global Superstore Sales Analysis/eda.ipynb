```python
import pandas as pd
```

```python
df = pd.read_csv('./data/superstore.csv')
print(df.shape)
df.columns
```

```python
df.info()
```

```python
#1. What is the overall sales, profit, and quantity sold
totals = df[['Sales', 'Profit', 'Quantity']].sum()

print(f"Total Sales: ${totals['Sales']:,}")
print(f"Total Profit: ${totals['Profit']:,}")
print(f"Total Quantity Sold: ${totals['Quantity']:,}")
```

```python
#2. What are the data types of each column?
df.dtypes
```

```python
#3. Convert Order.Date & Ship.Date to datetime.
df['Order.Date'] = pd.to_datetime(df['Order.Date'])
df['Ship.Date'] = pd.to_datetime(df['Ship.Date'])
```

```python
#4. What is the number of unique products, customers, and countries?
df.nunique()[['Product.ID', 'Customer.ID', 'Country']]
```

```python
#5. What are the top 5 selling product categories by sales?
df.groupby('Category')['Sales'].sum().sort_values(ascending=False).head()
```

```python
#6. What percentage of orders belong to each segment?
orders_by_segment = (df['Segment'].value_counts(normalize=True)*100)
orders_by_segment
```

```python
#7. Which 10 states have the highest number of orders?
df['State'].value_counts().head(10)
```

```python
#8. What is the average discount?
df['Discount'].mean()
```

```python
#9. Find the correlation between Sales, Profit, Quantity, Discount.
df[['Sales', 'Profit', 'Quantity', 'Discount']].corr()
```

```python
#10. Which ship mode is most frequently used?
df['Ship.Mode'].value_counts()
```

```python
#11. Show the first & last order dates.
df['Order.Date'].min(), df['Order.Date'].max()
```

```python
#12. What is the average shipping cost?
df['Shipping.Cost'].mean()
```

```python
#13. How many rows are there per year?
df['Year'].value_counts()
```

```python
#14. What are the top 5 most profitable sub-categories?
df.groupby('Sub.Category')['Profit'].sum().sort_values(ascending=False).head()
```
