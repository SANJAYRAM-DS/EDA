```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/LengthOfStay.csv')
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.head()
```

```python
df['vdate'] = pd.to_datetime(df['vdate'], errors='coerce')
```

```python
df['year'] = df['vdate'].dt.year
df['month'] = df['vdate'].dt.month
df['day'] = df['vdate'].dt.day
```

```python
avg_los = df['lengthofstay'].mean()
median_los = df['lengthofstay'].median()
max_los = df['lengthofstay'].max()
min_los = df['lengthofstay'].min()
```

```python
facility_counts = df['facid'].value_counts()
facility_counts.head()
```

```python
faci_avg_los = df.groupby('facid')['lengthofstay'].mean().sort_values(ascending=False)
faci_avg_los.head()
```

```python
plt.hist(df['lengthofstay'], bins=30, edgecolor='black')
plt.title("Distribution of Length of stay")
plt.xlabel("Length of Stay")
plt.ylabel("Frequency")
plt.show()
```

```python
p90 = df['lengthofstay'].quantile(0.90)
long_stay = df[df['lengthofstay'] > p90]

p90, long_stay.shape
```

```python
comorbidity_cols = ['dialysisrenalendstage', 'asthma', 'irondef', 'pneum', 
                    'substancedependence', 'psychologicaldisordermajor',
                    'depress', 'psychother', 'fibrosisandother', 'malnutrition']

df[comorbidity_cols].sum()
```

```python
df['clinical_risk'] = df[comorbidity_cols].sum(axis=1)
df[['eid', 'clinical_risk']].head()
```

```python
df[['clinical_risk', 'lengthofstay']].corr()
```

```python
sns.boxplot(x='clinical_risk', y='lengthofstay', data=df)
plt.title("LOS by ClinicalRisk")
plt.show()
```

```python
monthely_counts = df.groupby(['year', 'month'])['eid'].count()
monthely_counts.plot(kind='line', figsize=(10, 5))
plt.title("Monthly Admissions Trend")
plt.ylabel("Number of Admissions")
plt.show()
```
