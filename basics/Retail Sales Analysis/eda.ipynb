```python
import pandas as pd
```

#Cleaning and combaining datasets


```python
df_sales = pd.read_excel("data/mrtssales92-present.xls")   # Business-level sales by month
df_naics = pd.read_csv("data/2017 NAICS Definition.csv")   # NAICS code lookup
df_trend1 = pd.read_csv("data/SeriesReport-Seasonally Adjusted Sales - Monthly (Millions of Dollars).csv")  # Monthly trend (e.g., total retail sales)
df_trend2 = pd.read_csv("data/SeriesReport-Not Seasonally Adjusted Sales - Monthly (Millions of Dollars).csv")  # Monthly trend (e.g., revenue or inventory)
```

```python
for df in [df_sales, df_naics, df_trend1, df_trend2]:
    df.columns = (
        df.columns
        .str.strip()
        .str.replace(' ','_')
        .str.replace('  ','_')
        .str.replace('.','',regex=False)
    )
```
